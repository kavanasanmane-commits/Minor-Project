**Project Title**

Student marks analyzer using python and numpy


**problem statement**

To develop a student marks analyzer using python and numpy. The program takes marks of multiple students as input and calculates parameters such as total marks, average, highest mark, and lowest mark. Using numpy simplifies the calculation and improves efficiency when working with multiple values.

**Logic explanation**

1. Take input for the number of students (n) and number of subjects (s) — the   same for all students.

2. For each student, loop through all subjects and take marks as input, summing them to get the total marks for that student.

3. Calculate the average marks of each student by dividing the total by the number of subjects and store it in a list.

4. Convert the student averages list into a NumPy array for easier mathematical operations.

5. Use NumPy functions to find total of all students’ averages, overall average, highest average, and lowest average.

6. Assign a grade based on the overall average using conditional statements (if-elif-else).












```python

import numpy as np

n = int(input("Enter number of students: "))
subjects = int(input("Enter number of subjects (same for all students): "))

marks_list = []

for i in range(n):
    total_marks = 0
    print("Enter marks for student",i+1)

    for j in range(subjects):
        m = float(input(f"Enter the  student marks for subject{j+1}: "))
        total_marks = total_marks + m

    average_marks = total_marks / subjects
    marks_list.append(average_marks)

print("Student averages:", marks_list)

marks_array = np.array(marks_list)

total = np.sum(marks_array)
average = np.mean(marks_array)
highest = np.max(marks_array)
lowest = np.min(marks_array)

print("Total Marks:", total)
print("Average Marks:", average)
print("Highest Marks:", highest)
print("Lowest Marks:", lowest)

if average >= 85:
    grade = "A"
elif average >= 70:
    grade = "B"
elif average >= 50:
    grade = "C"
else:
    grade = "Fail"

print("Overall Grade:", grade)
```
